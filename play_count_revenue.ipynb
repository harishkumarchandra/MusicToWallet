```python
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
file_path = Path("play_count_v3.csv")
playcount_df = pd.read_csv(file_path)
playcount_df.tail()
```

```python
playcount_grp_df = playcount_df.groupby("popularity").mean()
```

```python
playcount_grp_df["play_count"] = playcount_grp_df["play_count"].astype(int)
playcount_grp_df["spotify_revenue"] = playcount_grp_df["spotify_revenue"].astype(int)
playcount_grp_df["apple_revenue"] = playcount_grp_df["apple_revenue"].astype(int)
```

```python
for_graph = playcount_grp_df[["spotify_revenue","apple_revenue"]]
```

```python
for_graph.plot.line(figsize=(12,8))
```
